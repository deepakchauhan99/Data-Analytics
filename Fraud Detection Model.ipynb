```python
import datetime
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split 
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
df= pd.read_csv("fraudTrain.csv")
df.head()
```

```python
df["trans_date_trans_time"] = pd.to_datetime(df["trans_date_trans_time"])
df["dob"] = pd.to_datetime(df["dob"])
```

```python
df.info()
```

```python
df.head(10)
```

```python
df.drop(columns = ["Unnamed: 0","cc_num","first","last","street","city","state","zip","dob","trans_num","trans_date_trans_time"],inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df
```

```python
merchant_encoder = LabelEncoder()
category_encoder = LabelEncoder()
gender_encoder = LabelEncoder()
job_encoder = LabelEncoder()

df["merchant"] = merchant_encoder.fit_transform(df["merchant"])
df["category"] = category_encoder.fit_transform(df["category"])
df["gender"] = gender_encoder.fit_transform(df["gender"])
df["job"] = job_encoder.fit_transform(df["job"])
```

```python
exit_counts = df["is_fraud"].value_counts().sort_index()
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.pie(exit_counts, labels=["No","Yes"],autopct="%0.3f%%")
plt.title("is_fraud Counts")
plt.tight_layout()
plt.show()
```

```python
df.info()
```

```python
X=df.drop(columns=["is_fraud"],inplace=False)
y=df["is_fraud"]
```

```python
model=SVC()
model.fit(X,y)
```

```python
model.score(X,y)
```
